`````python
def buscar_libro(libreria, titulo="ninguno"):
    
    if titulo == "ninguno":
        print("No se ha proporcionado ningún título para buscar")
        return
    
    encontrado = False
    
    for libro in libreria:
        if libro["titulo"] == titulo:
            print("¡LIBRO ENCONTRADO!")
            print("Título: ", (libro['titulo']))
            encontrado = True
            break
    
    if not encontrado:
        print(f"\nEse no lo tengo, ¿mola?")

buscar_libro(libreria, "1984")
`````

`````python
def validar_email(email):
    if "@" in email:
        return True
    else:
        return False

email_usuario = input("Ingresa tu dirección de email: ")

if validar_email(email_usuario):
    print("La dirección de email es VÁLIDA")
else:
    print("La dirección de email NO es válida (debe contener '@')")
`````
