## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
altura = input("Introduce la altura (mm):")
base = input("Introduce la base (mm):")
altura_1 = altura.replace(",", ".")
base_1 = base.replace(",", ".")
area = float(altura_1) * float(base_1)
print("Tu paralelogramo tiene un area de", round(area, 2) , "mm²")
```

## Ejercicio 5
Complete el siguiente código para solicitar al usuario su nombre y su edad. Luego, calcule en qué año nació el usuario y muestre el resultado.

```python
nombre = input("¿Cuál es tu nombre? ")
edad = input("¿Cuál es tu edad? ")
año_actual = 2026  # Puedes cambiarlo según el año en curso
año_nacimiento = int(año_actual) - int(edad)

print(nombre, "naciste en el año", año_nacimiento, "o el año", año_nacimiento - 1)
```
