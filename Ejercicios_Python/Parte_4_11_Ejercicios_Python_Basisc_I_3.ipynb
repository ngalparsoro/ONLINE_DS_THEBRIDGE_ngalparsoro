## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1
print(A == B)
print(A != C)
print(A > C)
print(C <= A)
print(B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
hermanos = input("Cuantos hermanos tienes?")
hijos = input("Cuantos hijos tienes?")
if hijos == hermanos:
    print("Tienes la misma cantidad de hermanos que hijos")
else:
    print("No tienes la misma cantidad de hermanos que hijos")
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
tios = input ("Cuantos tíos tienes?")
hermanos = input("Cuantos hermanos tienes?")
hijos = input("Cuantos hijos tienes?")
if hijos == hermanos == tios:
    print("Tienes la misma cantidad de todo")
elif hijos == hermanos or hermanos == tios or hijos == tios:
    print("Coinciden dos")
else:
    print("No coincide ninguno")
````
